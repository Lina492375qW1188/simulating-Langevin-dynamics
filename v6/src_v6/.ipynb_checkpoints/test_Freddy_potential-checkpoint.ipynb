```python
# Analyze Data
import numpy as np
import matplotlib.pyplot as plt
data=np.loadtxt("/Users/lina492375qw1188/Desktop/src_v4/xvyw1beta12.0gammax1.0gammay1.0epsln1.0sgma1.0A0.3x01.122w0.2B0.15a1.0_h0.002.txt", skiprows=1)
beta=15.0
x = data.T[0]
v = data.T[1]
y = data.T[2]
w = data.T[3]

plt.plot(x,y)

hist = np.histogram2d(x,y, bins=30)
plt.contour(hist[1][1::], hist[2][1::], (-1/beta)*np.log(hist[0].T+1e-7))
plt.title('Free Energy Surface')
plt.xlabel('x', size=16)
plt.ylabel('y', size=16).set_rotation(0)
plt.colorbar()
plt.show()
```

```python
def pe_Gaussian(x, y, x0, y0, iwx, iwy):
    return np.exp(-iwx*(x-x0)**2)*np.exp(-iwy*(y-y0)**2)
def u(x,y):
    return pe_Gaussian(x, y, 1.5, 1.8, 2, 2) + pe_Gaussian(x, y, 0.5, -0.5, 1, 1) + pe_Gaussian(x, y, -1.5, 1.5, 1, 1)
K=0.5*(v**2+w**2)
V=u(x,y)
plt.plot(K+V)
plt.ylabel("Etot")
plt.xlabel("step")
plt.show()
```
